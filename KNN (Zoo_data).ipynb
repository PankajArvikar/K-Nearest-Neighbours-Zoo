## Importing Libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import GridSearchCV,  train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import warnings 
warnings.filterwarnings('ignore')
```

## Importing Dataset

```python
zoo = pd.read_csv('Zoo.csv')
zoo.head()
```

## EDA

#### Dropping 'animal name' column which is not needed in the modelling

```python
zoo.drop('animal name', axis=1, inplace=True)
zoo.head()
```

### Overview of dataset

```python
zoo.isnull().sum()
```

```python
zoo.describe()
```

```python
zoo_array = zoo.values
```

### Splitting dataset into features & labels

```python
x = zoo_array[:,:-1]
y = zoo_array[:,-1]
```

### Splitting dataset into train-test

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=40)
```

## Building Model

```python
n_neighbors = np.array(range(2,40))
param_grid = dict(n_neighbors = n_neighbors)
```

```python
model = KNeighborsClassifier()
grid = GridSearchCV(estimator=model, param_grid=param_grid, cv= 10)
grid.fit(x_train, y_train)
```

```python
print(grid.best_score_)
print(grid.best_params_)
```

#### It shows that with 3 nearest neighbours, getting highest score.

### Checking accuracy of model on test data

```python
model_final = KNeighborsClassifier(n_neighbors=3)
model_final.fit(x_train, y_train)
```

```python
pred = model_final.predict(x_test)
accuracy_score(y_test, pred)
```
